```python
S = {1,2,3,4}
T = {3,4,5,6}


def intersection():
  intersection=[]
  for i in S:
    if i in T:
      intersection.append(i)
  return intersection    


intersection()
```

```python
Set = {-4, -2, 1, 2, 5, 0}
S=[]
for i in Set:
  S.append(i)

n= len(S)

def summation(S,n):
  found = False

  for i in range (0,n):
    for j in range (0,n):
      for k in range (0,n):
        if (S[i] + S[j] + S[k] is 0):
          my_tuple = (S[i], S[j], S[k])
          found = True
          print(my_tuple)

  if (found is False):
    print(" Not Existed ")        

summation(S,n)
```
